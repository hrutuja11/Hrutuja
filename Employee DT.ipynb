```python
for i in df:                  # i = column name 
    if df[i].dtypes in ("float64","int64"): # df[i] = every column will select float & int column only
        q1 = df[i].quantile(0.25)  # for 1 column compute q1
        q3 = df[i].quantile(0.75)  # for 1 column compute q3
        iqr = q3-q1  # for 1 column compute IQR
        ul = q3 + 1.5*iqr # for 1 column compute UPPER LIMIT
        ll = q1 - 1.5*iqr # for 1 column compute LOWER LIMIT 
        df[i].clip(lower=ll,upper=ul,inplace=True)
```

```python
import seaborn as sns
sns.boxplot(data=df)
```

```python
df.isna().sum()
```

```python
num=df.select_dtypes(include=[np.number])
num.head()
```

```python
cat=df.select_dtypes(include=['object'])
cat
```

```python
from sklearn.preprocessing import LabelEncoder

cat = cat.apply(LabelEncoder().fit_transform)
cat.head(3)
```

```python
combined = pd.concat([num, cat],axis=1)
combined.head()
```

```python
from sklearn.model_selection import train_test_split

x = combined.drop(['Attrition'],axis=1)
y = combined[['Attrition']]

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state=231)
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor

model = DecisionTreeClassifier() 
model.fit(X_train,y_train) 
```

```python
train = pd.concat([y_train,X_train],axis=1)
train.head(2)
```

```python
independent_variable = list(train.columns[1:])
independent_variable
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (5,4), dpi=300)
plot_tree(model,
               feature_names = independent_variable,  # column name
               class_names = ['No', 'Yes'], #  No,Yes
               filled = True, # colour
              node_ids=True, # node number
              fontsize=2); # 
#fig.savefig('imagename.png')
```

```python
train['Predicted'] = model.predict(X_train)  # MODEL = dt
train.head()
```

```python
from sklearn.metrics import confusion_matrix  
matrix = confusion_matrix(train['Predicted'],train['Attrition'])
print(matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(train['Attrition'], train['Predicted']))
```
